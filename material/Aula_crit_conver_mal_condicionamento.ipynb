Sejam os sistemas de equações lineares

\begin{aligned}
    3x_1 + x_2 + 7x_3     &= 12\\
    -2x_1 + 6x_2  -3x_3 &= 1 \\
    -5x_1 + 2x_2  + 2x_3 &= -1 \\

e 

 $\mathbf{R} : \left\{
\begin{aligned}
    x_1 + 4x_2 + x_3     &= 6\\
    -x_1 + x_2  -3x_3 &= -3 \\
    -2x_1 + x_2  + x_3 &= 0 \\
\end{aligned}
\right. $ 

ambos com solução exata $ \mathbf{x} = (1;1;1)$. Então:


*   (a) Verifique se o sistema $\mathbf{S}$ possui dominância por linha, ou por coluna. Calcule os autovalores da matriz **D** associada a configuação de dominância, caso ela ocorra, onde a matriz **D** é definida tal que **Ax = b** é equivalente a **x = Dx + c**.

*   (b) Faça mesma análise para o sistema Verifique se o sistema $\mathbf{R}$.



. 

*(a)* Em $\mathbf{S}$, troque as posições das linhas 1 e 3, o que resulta em: 

 $\mathbf{S'} : \left\{
\begin{aligned}
    -5x_1 + 2x_2  + 2x_3 &= -1 \\ 
    -2x_1 + 6x_2  -3x_3 &= 1 \\
    3x_1 + 2x_2 + 7x_3     &= 12\\
\end{aligned}
\right. $

*condição de dominânia por linha:* $|a_{ii}| > \sum_{j=1}^3|a_{ij}|$ com $j\neq i$, para $i=1,2,3$.

para linha 1: $|-5| > |2| + |2|$

para linha 2: $|6| > |-2| + |-3|$

para linha 3: $|7| > |3| + |2|$

Observe que a condição de dominância por coluna 
NÃO é satisfeita pois falha na coluna 1: $|-5| > |-2| + |3|$ . 

Portanto, o sistema $\mathbf{S'}$ é dominante, uma vez que possui dominância por linha!

Sejam: 

$\mathbf{D}_{\mathbf{S}} = \left[ 
  \begin{array}{c}
  0 & -1/3 &  -7/3\\
  1/3 & 0  & 1/2 \\
  -3/7 & -2/7 & 0 \end{array}
 \right]$ 

 e

 $\mathbf{D}_{\mathbf{S'}} = \left[ 
  \begin{array}{c}
  0 & 2/5 &  2/5\\
  1/3 & 0  & 1/2 \\
  5/2 & -1 & 0 \end{array}
 \right]$  

Os autovalores são calculados abaixo

```python
import numpy as np
n = 3
A_s = [[3, 1, 7],[-2, 6, -3], [-5, 2, 2]]  
A_s_linha = [[-5, 2, 2], [-2, 6, -3], [3, 1, 7]]                                
D_s = np.zeros((3,3))           # definicao de uma matriz de zeros
D_s_linha = np.zeros((3,3))           # definicao de uma matriz de zeros
for i in range(0,3,1):
  for j in range(0,3,1):
    if (i!=j):
      D_s[i][j] = -A_s[i][j]/A_s[i][i]
      D_s_linha[i][j] = -A_s_linha[i][j]/A_s_linha[i][i]

print('D_s = \n', np.matrix(D_s))
print('\nD_s_linha = \n', np.matrix(D_s_linha))

autovalores_D_s = np.linalg.eigvals(D_s)
print('\n\nautovalores de D_s = ',np.matrix(autovalores_D_s))
print('modulo dos autovalores de D_s = ',np.abs(autovalores_D_s))

autovalores_D_s_linha = np.linalg.eigvals(D_s_linha)
print('\n\nautovalores de D_s_linha = ',np.matrix(autovalores_D_s_linha))
print('modulo dos autovalores de D_s_linha = ',np.abs(autovalores_D_s_linha))
```

 As conclusões que podem ser tiradas *a priori*, são que:  

*   O sistema  $\mathbf{S}$, é diagonal dominante pois, basta realizar uma troca de linhas, gerando um sistema equivalente $\mathbf{S'}$, para que as condições de dominância sejam satisfeitas em $\mathbf{S'}$; 
*   Se executarmos um método iterativo no sistema original $\mathbf{S}$, não há garantias sobre convergência. 
*   Mas, se o mesmo processo iterativo for executado no sistema $\mathbf{S'}$, é garantido a convergência para uma solução aproximada, em um número finito de iterações. 


*(b)* Observe o sistema $\mathbf{R}$ com cuidado:

 $\mathbf{R} : \left\{
\begin{aligned}
    x_1 + 4x_2 + x_3     &= 6\\
    -x_1 + x_2  -3x_3 &= -3 \\
    -2x_1 + x_2  + x_3 &= 0 \\
\end{aligned}
\right. $ 

$\mathbf{R}$ não é dominante nem por linha, nem por coluna. 

Se tentarmos a troca por linha: 
* coloque $L_1$ na linha 2,  $L_2$ na linha 3, e $L_3$ na linha 1. Assim teremos: 

 $\mathbf{R'} : \left\{
\begin{aligned}
    -2x_1 + x_2  + x_3 &= 0 \\
    x_1 + 4x_2 + x_3     &= 6\\
    -x_1 + x_2  -3x_3 &= -3 \\
\end{aligned}
\right. $

* o teste de dominância falha para linha 1: $|-2| > |1| + |1|$ 


Se tentarmos a troca por coluna: 
* coloque $C_2$ na coluna 1,  $C_3$ na coluna 2, e $C_1$ na coluna 3. Assim teremos: 

 $\mathbf{R''} : \left\{
\begin{aligned}
    4x_2 + x_3 + x_1  &= 6\\
     x_2  - 3x_3 - x_1  &= -3 \\
     x_2  + x_3 -2x_1 &= 0 \\
\end{aligned}
\right. $ 

* o teste de dominância falha para coluna 3: $|-2| > |1| + |-1|$ 


As conclusões que podem ser tiradas *a priori*, são que:  

*   O sistema  $\mathbf{R}$, é não diagonal dominante pois mesmo ao realizar uma troca de linhas ou de colunas, gerando um sistema equivalente $\mathbf{R'}$, nenhuma das condições de dominância é satisfeita; 
*   Portanto, se e executarmos um método iterativo no sistema original $\mathbf{R}$, não há garantias sobre convergência. 




```python
import numpy as np
n = 3
A_r = [[1, 4, 1],[-1, 1, -3], [-2, 2, 2]]                                 
D_r = np.zeros((3,3))           # definicao de uma matriz de zeros
for i in range(0,3,1):
  for j in range(0,3,1):
    if (i!=j):
      D_r[i][j] = -A_r[i][j]/A_r[i][i]

print('D_r = \n', np.matrix(D_r))
autovalores_D_r = np.linalg.eigvals(D_r)
print('\n\nautovalores de D_s = ',np.matrix(autovalores_D_r))
print('modulo dos autovalores de D_s = ',np.abs(autovalores_D_r))


```

Como pode ser visto acima, pelo menos um dos autovalores de matriz $\mathbf{D}_{\mathbf{R}}$ é, em módulo, maior que 1, como o esperado, para um sistema sem dominância. 

# Aula 13 - Sistemas mal Condicionados

Sistemas mal condicionados são aqueles cuja a solução é "sensível" a pequenas mudanças em um ou algun(s) dos seus coeficientes. Sejam os sistemas $\mathbf{S}$ e $\mathbf{R}$: 

$\mathbf{S} :
\left\{
\begin{aligned}
    4.2x_1 + 3.4x_2     &= 0.8\\
    2.3x_1 - 1.2x_2 &  = 3.5
\end{aligned}
\right.$ 

com solução exata $ \mathbf{x} = (1;-1)$, e 

$\mathbf{R} :
\left\{
\begin{aligned}
    0,4071x_1 + 0.2434x_2     &= 0.4028\\
    0.2038x_1 +0.1218x_2 &= 0.2014
\end{aligned}
\right.$ 

com solução exata $ \mathbf{x} = (-2;5)$. Avalie se os sistemas $\mathbf{S}$ e $\mathbf{R}$ são mal condicionados. 

Fazendo uma pequena mudança no sistema $\mathbf{S}, 4.2 \rightarrow 4.21$  tem-se: 

$\mathbf{S'} :
\left\{
\begin{aligned}
    4.21x_1 + 3.4x_2     &= 0.8\\
    2.3x_1 - 1.2x_2 &  = 3.5
\end{aligned}
\right.$ 

 e a solução do sistema $\mathbf{S'}$ será:


```python
a = np.array([[4.21, 3.4], [2.3, -1.2]])
b = np.array([0.8, 3.5])
from scipy import linalg
y = linalg.solve(a, b)
y
```

Observe que $[ 0.99906774; -1.00178682]$ é muito proximo de $[1; -1]$, uma diferença menor que 2%. 


Fazendo uma pequena mudança  no sistema  $\mathbf{R}, 0.1218 \rightarrow 0.12185$, tem-se: 

$\mathbf{R'} :
\left\{
\begin{aligned}
    0,4071x_1 + 0.2434x_2     &= 0.4028\\
    0.2038x_1 +0.12185x_2 &= 0.2014
\end{aligned}
\right.$ 

e a solução do sistema $\mathbf{R'}$ será:


```python
aa = np.array([[0.4071, 0.2434], [0.2038, 0.12185]])
bb = np.array([0.4028, 0.2014])
from scipy import linalg
yy = linalg.solve(aa, bb)
yy

#from numpy.linalg import inv
#ainv = inv(aa)
#ainv
```

Compare as soluções $[2; 5]$ com $[281.02325581; -468.37209302]$, é mais de 90 vezes maior. Portanto este é um forte indicativo de mal condicionamento.

Para os sistemas $\mathbf{R}$ e $\mathbf{R'}$ da questão 2, com solução exata $(-2;5)$, execute o método de Gauss-Seidel, partindo das seguintes condições iniciais: $\mathbf{x}^{(0)} = (-1.5;4.75)$ e $\mathbf{x}^{(0)} = (-1.9;4.9)$, executando 250 iterações.  São portanto, quatro execuções. 

*(a) Sistema $\mathbf{R}$ e $\mathbf{x}^{(0)} = (-1.5;4.75)$:

```python
import numpy as np
n_int = 250
A = [[0.4071, 0.2434], [0.2038, 0.1218]]  # matriz dos coeficientes do sistema linear
b = [0.4028, 0.2014]                   # vetor com os termos independentes
D = np.zeros((2,2))
D[0][1] = -A[0][1]/A[0][0]
D[1][0] = -A[1][0]/A[1][1]
c = np.zeros((2,1))
c[0] = b[0]/A[0][0]
c[1] = b[1]/A[1][1]
p_min = 10**-8               # precisão mínima para parada
p_k = 1.                     # precisao obtida na iteracao k
p = [1., 1.]                 # vetor que armazena as precisoes na iteracao k
p_vetor = np.zeros(n_int)
k = 0
x_ant = [-1.5, 4.75]
x_pro = [0., 0.]
while (k<n_int)and(p_k>p_min):
  x_pro[0] = x_ant[1]*D[0][1] + c[0]
  x_pro[1] = x_pro[0]*D[1][0] + c[1]
  p[0] = np.fabs(x_pro[0] - x_ant[0])
  p[1] = np.fabs(x_pro[1] - x_ant[1])
  if (p[0] > p[1]):
     p_k = p[0]
  else:
     p_k = p[1]
  p_vetor[k] = p_k
  print('k = ', k+1, '  x_pro = ', x_pro[0],'  ',x_pro[1], '  p[0] = ', p[0], '  p[1] = ', p[1])
  x_ant[0] = x_pro[0]
  x_ant[1] = x_pro[1]
  k = k+1

import matplotlib.pyplot as plt  # biblioteca para plotar gráficos
plt.plot(range(0,k,1), p_vetor[0:k]) 
plt.grid()
plt.yscale('log')
plt.show()
```
